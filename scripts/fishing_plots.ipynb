```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
  
```

```python
# Plot between -10 and 10 with .001 steps.
x = np.arange(-4, 4, 0.01)
  
# Calculating mean and standard deviation
mean = statistics.mean(x)
sd = statistics.stdev(x)
  
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

cutoff = [0.35, 0.5, 0.65, 0.8]
xlen = len(x)
x1 = x[:int(xlen*cutoff[0])]
x2 = x[int(xlen*cutoff[0]):int(xlen*cutoff[1])]
x3 = x[int(xlen*cutoff[1]):int(xlen*cutoff[2])]
x4 = x[int(xlen*cutoff[2]):]
y1 = y[:int(xlen*cutoff[0])]
y2 = y[int(xlen*cutoff[0]):int(xlen*cutoff[1])]
y3 = y[int(xlen*cutoff[1]):int(xlen*cutoff[2])]
y4 = y[int(xlen*cutoff[2]):]


plt.fill_between(x1, y1, step="pre", alpha=0.4, color='b')
plt.fill_between(x2, y2, step="pre", alpha=0.4, color='purple')
plt.fill_between(x3, y3, step="pre", alpha=0.4, color='r')
plt.fill_between(x4, y4, step="pre", alpha=0.4, color='orange')


plt.show()
```

```python
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, ax = plt.subplots()

ax.set_xlim(-1, 1.2)
ax.set_ylim(-0.4, 0.4)

arr_1 = mpimg.imread('fishing_images/img_2.png')
arr_2 = mpimg.imread('fishing_images/img_1.png')
arr_3 = mpimg.imread('fishing_images/img_3.png')
arr_4 = mpimg.imread('fishing_images/img_0.png')

im1 = OffsetImage(arr_1, zoom=0.25)
x1 = AnnotationBbox(im1, (-0.75, 0.0))
im2 = OffsetImage(arr_2, zoom=0.25)
x2 = AnnotationBbox(im2, (-0.25, 0.0))
im3 = OffsetImage(arr_3, zoom=0.25)
x3 = AnnotationBbox(im3, (0.25, 0.0))
im4 = OffsetImage(arr_4, zoom=0.25)
x4 = AnnotationBbox(im4, (0.75, 0.0))

ax.add_artist(x1)
ax.add_artist(x2)
ax.add_artist(x3)
ax.add_artist(x4)

x_val = 0.0
plt.vlines(x_val, -0.25, ymax = 0.25, 
           color='b', 
           linestyles='dashed',
           label = 'First query mean')

x_val = 1.1
plt.vlines(x_val, -0.25, ymax = 0.25, 
           color='k', 
           linestyles='dashed',
           label = 'First query decision boundary')

#plt.grid()

plt.xlabel('Chosen feature')
ax.axes.yaxis.set_visible(False)
plt.draw()
plt.legend()
plt.show()
```
